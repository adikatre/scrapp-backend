```python
import os
import matplotlib.pyplot as plt
```

```python
path = "data/RecyclableAndHouseholdWasteClassification/images/"
categories= os.listdir(path)
print(categories)

category_names = []
image_counts = []
```

# Count the number of images in each category (subfolders)

```python
for category in os.listdir(path):
    category_path = os.path.join(path, category)
    
    count = 0

    # Loop through 'default' and 'real_world' folders
    for subfolder in ["default", "real_world"]:
        sub_path = os.path.join(category_path, subfolder)

        file_list = os.listdir(sub_path)
        png_files = []

        # Add each PNG file to the list
        for file in file_list:
            count += 1


    # Save results for plotting
    category_names.append(category)
    image_counts.append(count)

# Plot once after loop
plt.bar(category_names, image_counts)
plt.xticks(rotation=55, ha='right')
plt.xlabel("Category")
plt.ylabel("number of images")
plt.tight_layout()
plt.show()
```
