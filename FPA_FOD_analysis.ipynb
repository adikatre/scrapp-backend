# What counties are the most and least fire-prone?

Users should be able to enter in their latitude/longitude to see how fire-prone their area is based on historical data.

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("data/PA_FOD.sqlite")
df = pd.read_sql_query("SELECT NWCG_REPORTING_UNIT_NAME,FIRE_NAME,STAT_CAUSE_DESCR,LATITUDE,LONGITUDE,STATE,DISCOVERY_DATE FROM 'FIRES'", conn)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["LONGITUDE"], df["LATITUDE"], s = 0.5, alpha=0.01)
```

```python
df['STAT_CAUSE_DESCR'].value_counts().plot(kind='bar')
plt.show()
```
